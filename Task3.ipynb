#### **Importing Libraries**

```
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import sklearn.datasets as datasets 

#decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score,confusion_matrix
```

####**Loading Dataset**



```
iris=datasets.load_iris()
data=pd.DataFrame(iris.data, columns=iris.feature_names)
```

```
x = data
y=iris.target
from sklearn .model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 42)
```

#### **Decision Tree Algorithm**

```
dtree=DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

#### **Accuracy**

```
predict_tr = dtree.predict(x_train)
predict_te = dtree.predict(x_test)
```

```
print(confusion_matrix(predict_tr,y_train))
print(confusion_matrix(predict_te,y_test))
```

```
accuracy_train = accuracy_score(predict_tr,y_train)*100
accuracy_test = accuracy_score(predict_te,y_test)*100
print(accuracy_train)
print(accuracy_test)
```

```
sns.heatmap(confusion_matrix(y_test,predict_te),annot=True,cmap='YlGnBu')
plt.show()
```

```
plt.figure(figsize  = (20,20))
tree.plot_tree(dtree,filled = True,rounded = True,proportion = True,node_ids = True , feature_names = iris.feature_names)
plt.show()
```
